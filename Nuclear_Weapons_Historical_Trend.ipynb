# Nuclear weapon tests

## Year wise analysis

```python
tests.set_index("year", inplace = True)
```

```python
year_wise_tests = pd.read_csv("year_wise_tests.csv")
```

### "year_wise_tests.csv" file was generated out of the "nuclear_weapons_tests_states.csv" file using MySQL.
### Query used :- 
###  SELECT year, SUM(nuclear_weapons_tests) AS total_tests FROM nukes_test_states GROUP BY year;

```python
plt.figure(figsize = (10,6))
years = year_wise_tests["year"]
total_tests = year_wise_tests["total_tests"]
plt.plot(years, total_tests)
plt.title("Year wise nukes test")
plt.xlabel("Year", labelpad = 10)
plt.ylabel("Number of tests", labelpad = 10)
```

### Tests w.r.t Countries

```python
country_wise_test = pd.read_csv("country_wise_tests.csv")
```

### "country_wise_tests" is geberated from "nuclear_weapons_tests_states.csv" file using MySQL.
### Query used :- 
### SELECT country_name, SUM(nuclear_weapons_tests) FROM nukes_test_states GROUP BY country_name;

```python
country_wise_test.set_index("country_name", inplace = True)
country_wise_test["total_tests"].plot(kind = "bar")
```

```python
country_wise_test = pd.read_csv("country_wise_tests.csv")
total_tests = country_wise_test["total_tests"]
country_name = country_wise_test["country_name"]
my_labels = ["China", "France", "India", "North Korea", "Pakistan", "Russia", "UK", "US"]
colors = ["green", "blue", "yellow", "pink", "red", "orange", "grey", "skyblue"]
percent = 100.*(total_tests)/(total_tests.sum())
patches, texts = plt.pie(total_tests, colors = colors, startangle = 90, radius = 1.2, shadow = True)
labels = ["{0} - {1:1.2f} %".format(i, j) for i,j in zip(country_name, percent)]
plt.legend(patches, labels, loc = "best", bbox_to_anchor = (-0.1, 1.),
          fontsize = 10)
```

#### 2). Country wise nuclear weapon pursuits

```python
proliferations = pd.read_csv("proliferations.csv", index_col = 0)
```

#### "proliferations.csv" file was generated from "nuclear_weapons_proliferation_owid.csv" file using MySQL.
#### Query used :- 
#### SELECT country_name, SUM(nuclear_weapons_pursuit) AS Total_Pursuits FROM nuclear_weapons_proliferation
#### GROUP BY country_name HAVING Total_Pursuits != 0 ORDER BY 2;

```python
proliferations
```

```python
proliferations["Total_Pursuits"].plot(kind = "bar")
plt.xlabel("country_name", labelpad = 10, fontsize = 15)
plt.ylabel("number of pursuits", labelpad = 10, fontsize = 15)
plt.title("country wise nuclear weapon pursuits", fontsize = 15)
```

#### 3). Total nuclear weapon possession(country wise) since 1938

```python
total_possessions = pd.read_csv("total_possessions.csv")
```

#### "total_possessions.csv" file was generated from "nuclear_weapons_proliferation_owid.csv" file using MySQL
#### Query used :- 
#### SELECT country_name, SUM(nuclear_weapons_possession) AS total_uptill_now FROM nuclear_weapons_proliferation GROUP BY country_name HAVING total_uptill_now != 0 ORDER BY 2;

```python
total_possessions.set_index("country_name", inplace = True)
total_possessions
```

```python
total_possessions["total_uptill_now"].plot(kind = 'bar')
plt.ylabel("No. of Nukes", labelpad = 10, fontsize = 15)
plt.xlabel("Countries", labelpad = 10, fontsize = 15)
plt.title("Total possessions of nukes", fontsize = 15)
```
